```python
from sympy import randprime
from random import randint 
import time

def r_euclid(x, N):
        X = 1
        Y = 0
        R = 0
        S = 1
        while N != 0:
            C = x % N
            Q = x // N
            x, N = N, C
            R2, S2 = R, S
            R = X - (Q * R)
            S = Y - (Q * S)
            X, Y = R2, S2
        return x, X, Y

def fast_p(a, b, N):
    e = bin(b)
    c = 1
    for i in range(0, len(e)):
        c = (c ** 2) % N
        if e[i] == '1':
            c = (c * a) % N
    return c 
    
def gen(m):
    m1 = m // 2
    p = randprime(2^(m1-1), (2^m1)-1)
    q = randprime(2^(m1-1), (2^m1)-1)
    while p == q:
        q = randprime(2^(m1-1), (2^m1)-1)
    n = p * q
    phin = (p - 1) * (q - 1)
    while True:
        e = randint(3, phin - 1)
        eukl = r_euclid(e, phin)
        if eukl[0] == 1:
            d = eukl[1]
            break
    return (n, e), (n, d) #dlugosc modulu (n) jest rowna m 

def enc(pub, m):
    c = pow(m, pub[1], pub[0])
    return c

def dec(priv, c):
    m = pow(c, priv[1], priv[0])
    return m
        
def RSA(m):
    start = time.perf_counter()
    pub, priv = gen(m)
    end = time.perf_counter()
    n = pub[0]
    print("Klucz publiczny: (n, e) " + str(pub))
    print("Klucz prywatny: (n, d)" + str(priv))
    print("Czas dla generowania kluczy wynosi: " + str(end - start) + " s" + "\n")
    message = randint(3, n)
    print("Wiadomosc message = " + str(message) + "\n")
    start = time.perf_counter()
    c = enc(pub, message)
    end = time.perf_counter()
    print("Szyfrogram c = " + str(c))
    print("Czas dla szyfrowania wynosi: " + str(end - start) + " s" + "\n")
    start = time.perf_counter()
    message1 = dec(priv, c)
    end = time.perf_counter()
    if message1 == message:
        print("Deszyfrowanie sie powiodlo")
        print("Odzyskany tekst jawny: " + str(message1))
    else:
        print("Deszyfrowanie nie powiodlo sie")
    print("Czas dla deszyfrowania wynosi: " + str(end - start) + " s" + "\n")

def emulator():
    for i in range(100, 1100, 100):
        print("##########################################################")
        print(i)
        RSA(i)
emulator()
```
